```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
dataset = pd.read_csv("crime_data.csv")
dataset      
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)  
```

```python
# Normalization
```

```python
df_norm = norm_func(dataset.iloc[:,1:])
df_norm
  
```

```python
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
data = pd.DataFrame(trans.fit_transform(dataset.iloc[:,1:]))
data   
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
#p = np.array(df_norm) # converting into numpy array format 
z = linkage(df_norm, method="average",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')

plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=6.,  # rotates the x axis labels
    #leaf_font_size=15.,  # font size for the x axis labels
)
plt.show()     
```

```python
from sklearn.cluster import AgglomerativeClustering 
import warnings 
warnings.filterwarnings('ignore')
h_complete = AgglomerativeClustering(n_clusters=5, linkage='average',affinity = "euclidean").fit(df_norm) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
dataset['clust']=cluster_labels # creating a  new column and assigning it to new column 
dataset 
```

```python
dataset = pd.read_csv("crime_data.csv")
dataset  
```

```python
def norm_func(i):
    x = (i-i.min()) / (i.max() - i.min())
    return (x)  
```

```python
df_norm = norm_func(dataset.iloc[:,1:]) 
```

```python
df_norm.head(10)  # Top 10 rows
```

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()  
```

```python
clf = KMeans(n_clusters=4)
y_kmeans = clf.fit_predict(df_norm)  
```

```python
y_kmeans
#clf.cluster_centers_
clf.labels_ 
```

```python
y_kmeans 
```

```python
clf.cluster_centers_  
```

```python
clf.inertia_ 
```

```python
md=pd.Series(y_kmeans)  # converting numpy array into pandas series object 
dataset['clust']=md # creating a  new column and assigning it to new column 
dataset
```

```python
dataset.iloc[:,1:7].groupby(dataset.clust).mean() 
```

```python
#Import the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt    
```

```python
# Import crime .csv file and convert it to a DataFrame object
df = pd.read_csv("crime_data.csv")
print(df.head())
df  
```

```python
print(df.info())
```

```python
df1 = df.drop(['Unnamed: 0'],axis=1)    
```

```python
array=df1.values
array   
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array) 
X   
```

```python
dbscan = DBSCAN(eps=0.475, min_samples=5)
dbscan.fit(X)                      
```

```python
#Noisy samples are given the label -1.
dbscan.labels_          
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])   
```

```python
cl
pd.set_option("display.max_rows", None)  
```

```python
cl 
```

```python
df1 = pd.concat([df,cl],axis=1)  
df1     
```

```python
import matplotlib.pyplot as plt
plt.style.use('classic')   
```

```python
df1.plot(x="Assault",y ="Rape",c=dbscan.labels_ ,kind="scatter",s=50 ,cmap=plt.cm.copper_r) 
plt.title('Clusters using DBScan')      
```

```python
dl = dbscan.labels_ 
```

```python
from sklearn.cluster import KMeans
clf = KMeans(n_clusters=3)
y_kmeans = clf.fit_predict(X)  
```

```python
y_kmeans 
```

```python
cl1=pd.DataFrame(y_kmeans,columns=['Kcluster']) 
cl1 
```

```python
df2 = pd.concat([df1,cl1],axis=1) 
df2 
```

```python
df2.plot(x="Assault",y ="Rape",c=y_kmeans ,kind="scatter",s=50 ,cmap=plt.cm.copper_r) 
plt.title('Clusters using KMeans')   
```
